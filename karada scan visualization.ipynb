```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
url = 'https://raw.githubusercontent.com/rsydn9/karadascanvisual/main/karada.csv'
df = pd.read_csv(url, delimiter=';')
print(df)
```

```python
tgl = df['Tgl']
bb = df['BB (kg)']
fat = df['subs fat']
upper_fat = df['subs tran']
arm_fat = df['subs arm']
leg_fat = df['subs leg']
muscle = df['skel body']
upper_muscle = df['ske trans']
arm_muscle = df['ske arm']
leg_muscle = df['ske leg']
bf = df['bf %']
vis_fat = df['visc fat']
bmr = df['BMR']
bmi = df['BMI']
```

### Overall Body Composition

```python
fig, ax = plt.subplots()
plt.plot(tgl, fat, label='subcutaneous fat')
plt.plot(tgl, muscle, label='skeletal/muscle')
#plt.plot(tgl, vis_fat, label='visceral fat')
#plt.plot(tgl, bf %, label='body fat')
plt.xlabel('date')
plt.ylabel('body composition percentage')
plt.title("BODY RECOMPOSITION CHART")
plt.legend()
plt.show()
```

### Subcutaneous Fat

```python
fig, ax = plt.subplots()
plt.plot(tgl, upper_fat, label='subcutaneous trunk')
plt.plot(tgl, arm_fat, label='subcutaneous arm')
plt.plot(tgl, leg_fat, label='subcutaneous leg')
plt.xlabel('date')
plt.ylabel('sub fat percentage (%)')
plt.legend()
```

### Skeletal Mass/Muscle Mass

```python
fig, ax = plt.subplots()
plt.plot(tgl, upper_muscle, label='skeletal trunk')
plt.plot(tgl, arm_muscle, label='skeletal arm')
plt.plot(tgl, leg_muscle, label='skeletal leg')
plt.xlabel('date')
plt.ylabel('skeletal mass percentage (%)')
plt.legend()
```
